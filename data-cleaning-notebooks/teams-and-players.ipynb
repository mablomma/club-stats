# Teams and Players

```python
import json
import pandas as pd
```

### 1. List Teams for Club Icons

```python
f = open('./output/matches.json')
matches = json.load(f)
```

```python
teams = []
for m in matches:
    teams.append(m['home_team'])
    teams.append(m['away_team'])
teams = pd.DataFrame(teams).drop_duplicates(subset=['id']).sort_values(by=['id']).reset_index()[['id', 'name']]
```

```python
teams
```

### 2. List Players for Player Images

```python
f = open('./output/stats.json')
players = json.load(f)
```

```python
players = pd.DataFrame(players).drop_duplicates(subset=['player_id']).sort_values(by=['player_id'])
players = players[players['team_id'] == 977].reset_index()[['player_id', 'player_name']]
```
